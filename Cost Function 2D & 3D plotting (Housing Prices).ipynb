**Unilateral and bilateral derivation of cost function against changing values of model parameters w (weight) & b (bias)**

**- In this exercise we are taking input of changing housing prices with the change in sq foot size of the house**

```python
# NumPy, a popular library for scientific computing
# Matplotlib, a popular library for plotting data
import numpy as np
import matplotlib.pyplot as plt
```

**Define training arrays**

```python
x_train=np.array([1.0,2.0])  # size in 1000 sq ft
y_train=np.array([300.0,500.0]) # price in 1000s of $
```

**Write Compute Cost Function to predict how accurately model fits the data**

```python
def compute_cost_function(x,y,w,b):
    m=x.shape[0]
    sq_sum=0
    for i in range(m):
        sq_sum+=np.square((w*x[i]+b)-y[i])
    cost_j_wb=sq_sum/(2*m)

    return cost_j_wb
```

**Plot cost function derivation against changing value of weight (w)**

```python
b=100 # we will keep b fixed to calculate unilateral change for now

arr_w=np.array([-600,-500,-400,-300,-200,-100,0,100,200,300,400,500,600]) # we will calculate values of cost function with changing values in w
arr_w_size=arr_w.shape[0]
arr_j_w=np.zeros(arr_w_size)
for i in range(arr_w_size):
    arr_j_w[i]=compute_cost_function(x_train,y_train,arr_w[i],b)
    print(f"When value of w is {arr_w[i]} then Value of cost function is: {arr_j_w[i]}")

plt.plot(arr_w,arr_j_w,marker='o',c='r',label='Cost Prediction')
plt.xlabel("weight parameter change")
plt.ylabel("cost function change")
plt.title("Cost function change with changing w and b is 100")
plt.show()
```

**Plot cost function deviation against changing values of w and b in 2D**

```python
x_train = np.array([1.0, 1.7, 2.0, 2.5, 3.0, 3.2])
y_train = np.array([250, 300, 480,  430,   630, 730,])
arr_w=np.array([-500,-400,-300,-200,-100,0,100,200,300,400,500])
arr_b=np.array([500,400,300,200,100,0,-100,-200,-300,-400,-500])
arr_w_size=arr_w.shape[0]
arr_b_size=arr_b.shape[0]
arr_j_wb=np.zeros((arr_w_size, arr_b_size))

for i in range(arr_w_size):
    for j in range(arr_b_size):
        arr_j_wb[i,j]=compute_cost_function(x_train,y_train,arr_w[i],arr_b[j])

plt.contourf(arr_w,arr_b,arr_j_wb,levels=20,cmap='viridis')
plt.xlabel("weight parameter change")
plt.ylabel("bias parameter change")
plt.title("Cost function change with changing values of w & b")
plt.show()
```

**Plot cost function deviation with changing values of w and b in 3D**

```python
from mpl_toolkits.mplot3d import Axes3D

x_train = np.array([1.0, 1.7, 2.0, 2.5, 3.0, 3.2])
y_train = np.array([250, 300, 480,  430,   630, 730,])
arr_w=np.array([-500,-400,-300,-200,-100,0,100,200,300,400,500])
arr_b=np.array([500,400,300,200,100,0,-100,-200,-300,-400,-500])
arr_w_size=arr_w.shape[0]
arr_b_size=arr_b.shape[0]
arr_j_wb=np.zeros((arr_w_size, arr_b_size))

for i in range(arr_w_size):
    for j in range(arr_b_size):
        arr_j_wb[i,j]=compute_cost_function(x_train,y_train,arr_w[i],arr_b[j])

# Create a meshgrid for w and b
W, B = np.meshgrid(arr_w, arr_b)

# Create a 3D figure with a subplot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
surf = ax.plot_surface(W, B, arr_j_wb, cmap='viridis', rstride=1, cstride=1, alpha=0.8, linewidth=0, antialiased=False)

# Add labels and a title
ax.set_xlabel("Weight parameter (w)")
ax.set_ylabel("Bias parameter (b)")
ax.set_zlabel("Cost")
plt.title("3D Surface Plot of Cost Function")

# Add a color bar
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```
